### 1. O que √© Seaborn e por que us√°-lo?

O Seaborn √© uma biblioteca de visualiza√ß√£o baseada no Matplotlib que oferece:

- Interface de alto n√≠vel para gr√°ficos estat√≠sticos.

- Temas atraentes por padr√£o.

- Integra√ß√£o perfeita com DataFrames pandas.

- Visualiza√ß√µes complexas com poucas linhas de c√≥digo.

```python
# Importando as bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Verificando a vers√£o do Seaborn
print(f"Seaborn version: {sns.__version__}")


print("Bibliotecas importadas com sucesso!!")

# Vamos usar um dataset que j√° vem com o Seaborn
dados = sns.load_dataset("penguins")
```

### Compara√ß√£o R√°pida:

**Matplotlib**: Mais controle, mais c√≥digo

**Seaborn**: Menos c√≥digo, visualiza√ß√µes estat√≠sticas prontas

Vamos ver como s√£o nossos dados:

```python
# Primeira olhada nos dados
print("Primeiras linhas:")
print(dados.head())

print("\nFormato dos dados:")
print(dados.shape)
```

### üìã O que √© o Dataset Penguins?

O dataset Penguins (Pinguins) √© um conjunto de dados sobre pinguins que substituiu o famoso dataset Iris como exemplo padr√£o para iniciantes em data science. Ele cont√©m medi√ß√µes f√≠sicas de pinguins coletadas na Ant√°rtica.

### 2. Nosso Primeiro Gr√°fico: Scatter Plot

O scatter plot mostra a rela√ß√£o entre duas coisas.



```python
# Gr√°fico mais simples poss√≠vel
sns.scatterplot(data=dados, x='bill_length_mm', y='bill_depth_mm')
plt.show()
```

Agora vamos deixar mais interessante:

```python
# Colorindo por categoria
sns.scatterplot(data=dados, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Comprimento vs Profundidade do Bico por Esp√©cie')
plt.show()
```

### Exerc√≠cio 1:

Fa√ßa um scatter plot mostrando o peso corporal (body_mass_g) vs comprimento do bico (bill_length_mm), colorindo por ilha (island).

### 3. Gr√°fico de Barras

Perfeito para comparar grupos.

```python
# Gr√°fico de barras simples
sns.barplot(data=dados, x='species', y='body_mass_g')
plt.title('Peso M√©dio por Esp√©cie')
plt.show()
```

Com mais informa√ß√µes:

```python
# Comparando duas coisas ao mesmo tempo
sns.barplot(data=dados, x='species', y='body_mass_g', hue='sex')
plt.title('Peso M√©dio por Esp√©cie e Sexo')
plt.legend(title='Sexo')
plt.show()
```

### Exerc√≠cio 2:

Crie um gr√°fico de barras mostrando o comprimento m√©dio do bico por ilha.

### 4. Histograma

Para ver como os dados se distribuem.

```python
# Histograma simples
sns.histplot(data=dados, x='body_mass_g')
plt.title('Distribui√ß√£o do Peso Corporal')
plt.show()
```

Com mais detalhes:

```python
# Colorindo por categoria
sns.histplot(data=dados, x='body_mass_g', hue='species', alpha=0.5)
plt.title('Distribui√ß√£o do Peso por Esp√©cie')
plt.show()
```

### Exerc√≠cio 3:

Fa√ßa um histograma do comprimento do bico, colorindo por sexo.

### 5. Boxplot

Para ver a distribui√ß√£o e valores extremos.

```python
# Boxplot b√°sico
sns.boxplot(data=dados, x='species', y='body_mass_g')
plt.title('Distribui√ß√£o do Peso por Esp√©cie')
plt.show()
```

Com cores:



```python
# Boxplot colorido
sns.boxplot(data=dados, x='species', y='body_mass_g', hue='sex')
plt.title('Distribui√ß√£o do Peso por Esp√©cie e Sexo')
plt.legend(title='Sexo')
plt.show()
```

### 6. Mudando o Estilo

O Seaborn vem com temas prontos!

```python
# Experimente mudar o estilo:
sns.set_style("whitegrid")  # Tente: "darkgrid", "white", "dark", "ticks"

sns.scatterplot(data=dados, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.show()
```

Mudando cores:

```python
# Usando paletas de cores diferentes
sns.scatterplot(data=dados, x='bill_length_mm', y='bill_depth_mm', 
                hue='species', palette='Set2')
plt.show()
```
