### 1. O que é Matplotlib?

O Matplotlib é uma biblioteca Python para criação de visualizações estáticas, animadas e interativas. É uma das ferramentas mais populares para visualização de dados!

```python
# Importando a biblioteca
import matplotlib.pyplot as plt
print("Biblioteca matplotlib.pyplot importada com sucesso!")
```

Por que visualizar dados?

- Identificar padrões rapidamente

- Comunicar informações de forma clara

- Detectar anomalias ou outliers

- Tomar melhores decisões baseadas em dados

### 2. Nosso Primeiro Gráfico: Gráfico de Linha
Vamos começar com algo simples - um gráfico de linha!

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados simples
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criando o gráfico
plt.plot(x, y)

# Mostrando o gráfico
plt.show()
```

### 3. Customizando Nosso Gráfico

Um gráfico sem informações não é muito útil. Vamos adicionar alguns elementos:



```python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criando gráfico com customizações
plt.plot(x, y, color='red', linewidth=2, marker='o', markersize=8, label='Crescimento')

# Adicionando informações
plt.title('Meu Primeiro Gráfico Customizado')
plt.xlabel('Eixo X - Tempo')
plt.ylabel('Eixo Y - Valores')
plt.legend()

# Grade para melhor leitura
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

*Parâmetros Úteis*:

- color: cor da linha ('red', 'blue', '#FF5733')

- linewidth: espessura da linha

- marker: estilo do ponto ('o', 's', '^', 'D')

- linestyle: estilo da linha ('-', '--', '-.', ':')

### 4. Gráfico de Barras

Perfeito para comparar categorias!

```python
# Dados de exemplo: Vendas por mês
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai']
vendas = [150, 200, 180, 300, 250]

plt.bar(meses, vendas, color=['#FF9999', '#66B2FF', '#99FF99', '#FFD700', '#FF69B4'])
plt.title('Vendas Mensais')
plt.xlabel('Meses')
plt.ylabel('Vendas (unidades)')

# Adicionando valores em cima das barras
for i, v in enumerate(vendas):
    plt.text(i, v + 5, str(v), ha='center')

plt.show()
```

### 5. Gráfico de Dispersão (Scatter Plot)

Ideal para ver relações entre duas variáveis!

```python
# Gerando dados aleatórios
np.random.seed(42)  # Para resultados consistentes
x = np.random.rand(50) * 100
y = x * 2 + np.random.rand(50) * 30

plt.scatter(x, y, alpha=0.6, c='green', s=50)
plt.title('Relação entre X e Y')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.grid(True, alpha=0.3)
plt.show()

# Desafio: O que acontece se você mudar o valor de np.random.seed()?
```

### 6. Múltiplos Gráficos (Subplots)

Às vezes queremos comparar vários gráficos lado a lado:

```python
# Criando figura com subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Gráfico 1 - Linha
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

ax1.plot(x, y1, 'b-', label='sen(x)')
ax1.plot(x, y2, 'r--', label='cos(x)')
ax1.set_title('Funções Trigonométricas')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid(True)

# Gráfico 2 - Barras
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]
cores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

ax2.bar(categorias, valores, color=cores)
ax2.set_title('Dados por Categoria')
ax2.set_xlabel('Categorias')
ax2.set_ylabel('Valores')

# Ajustando layout
plt.tight_layout()
plt.show()
```

### 7. Salvando Seu Gráfico

Não esqueça de salvar seu trabalho!

```python
# Criando um gráfico para salvar
x = [1, 2, 3, 4, 5]
y = [x**2 for x in x]  # Quadrado de x

plt.plot(x, y, 'go-', linewidth=2, markersize=8)
plt.title('Gráfico de x²')
plt.xlabel('x')
plt.ylabel('x²')

# Salvando em diferentes formatos
plt.savefig('meu_grafico.png', dpi=300, bbox_inches='tight')
plt.savefig('meu_grafico.pdf')

plt.show()
```
